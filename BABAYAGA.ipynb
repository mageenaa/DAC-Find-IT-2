```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = df.dropna()
```

```python
str_cols = df.select_dtypes(include = 'object').columns
```

```python
df[str_cols].head()
```

```python
def prep(df):
   for col in df.columns:
        if df[col].isnull().sum() and df[col].dtypes == 'object':
            df[col].loc[(df[col].isnull())] = df[col].mode().max()
```

```python
df.isnull().sum()
```

```python
df_copy = df.copy()
```

```python
prep(df_copy)
```

```python
df_copy.isnull().sum()
```

```python
le = LabelEncoder()
for col in str_cols:
    df_copy[col] = le.fit_transform(df_copy[col])
```

```python
x = df_copy.drop('Price', axis=1)
y = df_copy['Price']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, classification_report
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier, StackingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
pipelines = {
    'rf' : make_pipeline(StandardScaler(), RandomForestClassifier(random_state=42)),
    'gb' : make_pipeline(StandardScaler(), GradientBoostingClassifier(random_state=42))
}
```

```python
grid = {
    'rf' : {
        'randomforestclassifier__n_estimators' : [100,200,300]
    },
    'gb' : {
        'gradientboostingclassifier__n_estimators' : [100,200,300]
    }
}
```

```python
fit_models = {}
for algo, pipeline in pipelines.items():
    print(f'Training {algo} model')
    model = GridSearchCV(pipeline, grid[algo], n_jobs=-1, cv=10)
    model.fit(x_train, y_train)
    fit_models[algo] = model 
```

# Ensemble Using Voting Classifier

```python
lr = LogisticRegression(random_state=42, max_iter=200)
knn = KNeighborsClassifier(n_neighbors=5)
dtc = DecisionTreeClassifier(random_state=42)
gbc = GradientBoostingClassifier(n_estimators=100, random_state=42)
rfc = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
svm = SVC(kernel='linear', probability=True, random_state=42)
```

```python
voting_hard = VotingClassifier(estimators=[('lr', lr),('knn', knn), ('dtc', dtc), ('gbc', gbc), ('rfc', rfc), ('svm', svm)], voting='hard')
voting_soft = VotingClassifier(estimators=[('lr', lr),('knn', knn), ('dtc', dtc), ('gbc', gbc), ('rfc', rfc), ('svm', svm)], voting='soft')
```

```python
voting_hard.fit(x_train, y_train)
voting_soft.fit(x_train, y_train)
```

```python
pred_hard = voting_hard.predict(x_test)
score_hard = mean_absolute_error(y_test, pred_hard)
print(f'Voting Hard MAE : ', score_hard)
```

```python
pred_soft = voting_soft.predict(x_test)
score_soft = mean_absolute_error(y_test, pred_soft)
print(f'Voting Soft MAE : ', score_soft)
```

```python
param_grid_xgboost = {
    "xgbclassifier__max_depth": [3, 4, 5, 7],
    "xgbclassifier__learning_rate": [0.1, 0.01, 0.05],
    "xgbclassifier__gamma": [0, 0.25, 1],
    "xgbclassifier__reg_lambda": [0, 1, 10],
    "xgbclassifier__scale_pos_weight": [1, 3, 5],
    "xgbclassifier__subsample": [0.8],
    "xgbclassifier__colsample_bytree": [0.5],
}
```
